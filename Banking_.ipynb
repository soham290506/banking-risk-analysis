```python
import pandas as pd

df = pd.read_csv("master_banking_updated.c.csv")
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Total_Loan'] = (
    df['Bank Loans'] +
    df['Credit Card Balance'] +
    df['Business Lending']
)
```

```python
df['Loan_to_Income'] = df['Total_Loan'] / df['Estimated Income']
```

```python
def risk_category(x):
    if x > 0.6:
        return "High Risk"
    elif x > 0.3:
        return "Medium Risk"
    else:
        return "Low Risk"

df['Risk Category'] = df['Loan_to_Income'].apply(risk_category)
```

```python
df['Joined Bank'] = pd.to_datetime(df['Joined Bank'], errors='coerce')

df['Engagement Days'] = (
    pd.Timestamp.today() - df['Joined Bank']
).dt.days
```

```python
def engagement_bucket(days):
    if days < 1825:
        return "<5 Years"
    elif days < 3650:
        return "<10 Years"
    else:
        return ">10 Years"

df['Engagement Bucket'] = df['Engagement Days'].apply(engagement_bucket)
```

```python
def income_band(x):
    if x < 100000:
        return "Low"
    elif x < 300000:
        return "Mid"
    else:
        return "High"

df['Income Band'] = df['Estimated Income'].apply(income_band)
```
